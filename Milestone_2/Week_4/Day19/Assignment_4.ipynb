Objective:

Understand and implement model evaluation using cross-validation and improve model performance by hyperparameter tuning.

Step 1: Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df=pd.read_csv('/content/sample_data/preprocessed_earthquake_data.csv')
```

Step 2: Load Dataset and Prepare Features and Target

```python
target ='Magnitude'
categorical_cols =['Magnitude','Type','Magnitude Type','Source','Status']

X =df.drop(columns=[target]+categorical_cols)
y =df[target]

model=RandomForestRegressor(random_state=42)
```

Step 3: Implement Cross-Validation

```python
scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
```

```python
print(f"Negative MAE for each fold:{scores}")
print(f"Mean :{np.mean(scores):.4f}")
print(f"Variance:{np.var(scores):.6f}")
```

Step 4: Hyperparameter Tuning with GridSearchCV

```python
param_grid ={
    'n_estimators':[50, 100, 200],
    'max_depth':[None, 10, 20],
    'min_samples_split':[2, 5]
}

grid_searches =GridSearchCV(estimator=model,param_grid=param_grid,cv=5,scoring='neg_mean_absolute_error')
grid_searches.fit(X,y)
```

Step 5: Evaluate Best Model on Full Dataset

```python
print(f"Best parameters={grid_searches.best_params_}")
print(f"Best cross-validation accuracy={grid_searches.best_score_:.4f}")
```
