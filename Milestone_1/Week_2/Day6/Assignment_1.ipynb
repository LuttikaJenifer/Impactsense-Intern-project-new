## Practice on Pandas and Data Visualization

1. What is the overall survival rate of passengers aboard the Titanic?

```python
df_1=df[['Survived']]
```

```python
Survived_rate = df_1.mean()*100
print(Survived_rate)
```

```python
df_2=df[['Survived','Pclass']]
```

2. How does the survival rate vary by passenger class (Pclass)?

```python
survived_rate_by_class = df_2.groupby('Pclass')['Survived'].mean()*100
print(survived_rate_by_class)
```

3. What is the average age of passengers who survived vs. those who did not survive?

```python
avg_age_survived = df[df['Survived'] == 1]['Age'].mean()
avg_age_not_survived = df[df['Survived'] == 0]['Age'].mean()
print(avg_age_survived)
print(avg_age_not_survived)
```

4. How many passengers embarked from each port (Embarked), and what are their survival rates?

```python
passengers_embarked = df['Embarked'].value_counts()
Survived_rate_embarked = df.groupby('Embarked')['Survived'].mean()*100
print(passengers_embarked)
print(Survived_rate_embarked)
```

5. What is the distribution of fare prices paid by passengers, and does fare impact survival chances?

```python
df["Fare"].describe()
```

```python
avg_fare = df.groupby("Survived")['Fare'].mean()
print(avg_fare)
```

6. Plot the count of passengers by passenger class (Pclass).

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Pclass')
plt.title('Count of passengers by passenger class')
plt.xlabel('Pclass')
plt.ylabel('count')
```

7. Visualize the survival rate with respect to gender using a bar chart.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Sex',hue='Survived')
plt.title('Survival rate with respect to gender')
plt.xlabel('Sex')
plt.ylabel('Count')
```

8. Show the age distribution of passengers using a histogram or KDE plot.

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='Age',kde=)
plt.title('Age distribution of passengers')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(data=df,x='Age',fill='True')
plt.title('Age distribution of passengers')
plt.xlabel('Age')
plt.ylabel('Count')
```

9. Create a boxplot comparing fares across different passenger classes.

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df,x='Pclass',y='Fare')
plt.title('Fare distribution across different Passenger class')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.ylim(0,300)
```

10. Visualize survival count by embarkation port (Embarked) with a stacked bar chart.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Embarked',hue='Survived')
plt.title('Survival count by embarkation port')
plt.xlabel('Embarked')
plt.ylabel('Count')
```

11. Display a heatmap of the correlation matrix of numerical features in the dataset.

```python
corr.mat=df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr.mat,annot=True,cmap='cool')
plt.title('Correlation matrix of numerical features')
```
