###MERGING AND JOINING

Default value of axis is 0

```python
df_1=temp_df['Latitude']
df_2=temp_df['Longitude']
df_3=pd.concat([df_1,df_2]) #concatenate vertically axis=0
df_3
```

```python
df_3=pd.concat([df_1,df_2],axis=1) # concatenate horizondally when axis=1
df_3
```

```python
df_1=temp_df[['Latitude','Magnitude']]
df_2=temp_df['Longitude']
df_3=pd.concat([df_1,df_2],axis=1)
df_3
```

```python
df_1=temp_df[['Latitude','Magnitude']]
df_2=temp_df[['Longitude','Root Mean Square']]
df_3=pd.merge(df_1,df_2,left_index=True,right_index=True) #Merge the indices of df_1 and df_2
df_3
```

### GROUP AND AGGREGATION

```python
temp_df.columns
```

```python
temp_df.groupby('Status').mean()
```

```python
temp_df.groupby('Status')[['Latitude','Magnitude']].mean()
```

```python
grouped=temp_df.groupby('Status').agg({'Magnitude':'mean','Root Mean Square':'sum'})
grouped
```

```python
result=temp_df.groupby('Status').agg({
    'Magnitude':['mean','min'],
    'Root Mean Square':['sum','max']
})
result
```

```python
def prod_two(group):
  return (group['Magnitude']*group['Root Mean Square']).sum()/group['Root Mean Square'].sum()

result=temp_df.groupby('Status').apply(prod_two)
print(result)
```
